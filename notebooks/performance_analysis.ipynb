```python
from pathlib import Path
import os
BASE_DIR = Path(os.getcwd(), "outputs_final")
```

```python
SCORES = {}
```

```python

# Ensemble 
import pandas as pd

pd.options.display.float_format = '{:,.1f}'.format


all_scores = []
for fold in range(10):
    #try: 
        subdir = BASE_DIR / f"ensemble_fold_nocal{fold}"
        subdir = subdir / subdir.iterdir().__next__()
        metrics_ensemble = pd.read_csv(subdir / "metrics_ensemble_no_tc.csv", index_col=0)
        metrics = {}
        for name, row in metrics_ensemble.iterrows():
            for set, item in dict(row).items():
                metrics[f'{set}_{name}'] = item

        metrics['fold'] = fold
        all_scores.append(metrics)
    #except: 
    #    print(f"Failed on fold {fold}")
    #    continue

all_scores
all_scores = pd.DataFrame(all_scores)
all_scores = all_scores.set_index('fold')

columns_of_interest = ['test_corewise_auc', 'test_corewise_sensitivity', 'test_corewise_specificity', 'test_patchwise_auc', 'test_patchwise_ece', 'test_patchwise_brier_score', ]
all_scores = all_scores.loc[:, columns_of_interest]
all_scores *= 100
SCORES['ensemble'] = all_scores
all_scores.describe()
```
